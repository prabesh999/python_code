```python
#interpolation formula
# importing data
file1= pd.read_csv('far_left.csv')
j=0
k=0
#finding the value greater and lesser than 0.88 for interpolation
data1=file1['norm']
for i in range(len(data1)):
    if data1[i]>= 0.88:  # 88%
        j+=1


x=0.88
x1=file1['norm'][j-1]
y1=file1['distance'][j-1]

x2= file1['norm'][j]
y2=file1['distance'][j]

#interpolation formula
#y_88 is the beamwith of laser when 88% of intensity is reduced
y_88= y1+(x-x1)*((y2-y1)/(x2-x1))


#finding the value greater and lesser than 0.12 for interpolation
data1=file1['norm']
for i in range(len(data1)):
    if data1[i]>= 0.12:  # 12%
        k+=1


X=0.12
X1=file1['norm'][k-1]
Y1=file1['distance'][k-1]

X2= file1['norm'][k]
Y2=file1['distance'][k]

#interpolation formula
# y is the beamwidth of laser when concave lens is move far left
y_12= Y1+(X-X1)*((Y2-Y1)/(X2-X1))

# y is the beamwidth of laser when concave lens is move far left
y_fl =(y_12-y_88)*10**-3 # changing the unit into mm from um. 
print(y_fl,'mm is the FWHM of beamwith when the concave lens is kept far left')
   
 #laserbeamwidth according to 13.5%
l=0
#finding the value greater and lesser than 0.88 for interpolation
data1=file1['norm']
for t in range(len(data1)):
    if data1[t]>= 0.135:  # 13.5%
        l+=1


x_e=0.135
x1_e=file1['norm'][l-1]
y1_e=file1['distance'][l-1]

x2_e= file1['norm'][l]
y2_e=file1['distance'][l]

#interpolation formula
#y_88 is the beamwith of laser when 88% of intensity is reduced
y_e= y1_e+(x_e-x1_e)*((y2_e-y1_e)/(x2_e-x1_e))
print(y_e*10**-3,'mm is the beamwidth when intenisty is reduced to 13.5%')
print(0.85*y_fl)
                                                    
```
