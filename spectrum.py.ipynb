```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
file1=pd.read_csv('Spectrum_20211208.sps')
file2=pd.read_csv('Spectrum_20211216.sps')
file3=pd.read_csv('Spectrum_20211216_b.sps')
file4=pd.read_csv('Spectrum_20211220_b.sps')
file5=pd.read_csv('Spectrum_20211220.sps')
file6=pd.read_csv('Spectrum_20211221_c.sps')
```

```python
#filtering the data

d1= file1["SpectraSmart Spectrum 2.6.0.19980"].str.split( " ", n = 1,  expand=True)[25:].astype('float')
d2=file2["SpectraSmart Spectrum 2.6.0.19980"].str.split( " ", n = 1,  expand=True)[25:].astype('float')
d3=file3["SpectraSmart Spectrum 2.6.0.19980"].str.split( " ", n = 1,  expand=True)[25:].astype('float')
d4= file4["SpectraSmart Spectrum 2.6.0.19980"].str.split( " ", n = 1,  expand=True)[25:].astype('float')
d5= file5["SpectraSmart Spectrum 2.6.0.19980"].str.split( " ", n = 1,  expand=True)[25:].astype('float')
d6= file6["SpectraSmart Spectrum 2.6.0.19980"].str.split( " ", n = 1,  expand=True)[25:].astype('float')

```

```python
#plot the graph

plt.plot(d1[0],d1[1])


plt.title('spectrum of white light without filter',fontweight='bold')
plt.xlabel('''Wavelength (nm)

           continuum pumped by 1030 nm and pulse energy of 1.7 $\mu$J  ''')
plt.ylabel('Intensity (a.u.)')
plt.legend(loc='best')
plt.savefig("spectrum_WLG_NF.png",dpi=200)
```
